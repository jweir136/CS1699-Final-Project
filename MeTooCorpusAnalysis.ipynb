# Get the author's categories from the hate speech terms

```python
categories = []
```

```python
for rating in tqdm(me_too_corpus['response_category'].values):
    categories.append(int(rating[0]))
```

```python
with open("curry_response_results.csv", "w+") as f:
    f.write("response\n")
    for result in categories:
        f.write("{}\n".format(result))
```
